```python
import numpy as np
import pandas as pd
```

```python
df_driver=pd.read_csv(r'drivers.csv')
df_driver
```

```python
driver=df_driver.iloc[:,0:2]
driver
```

```python
df_race=pd.read_csv(r"races.csv")
df_race
```

```python
race=df_race[['raceId','name','year']]
```

```python
df_result=pd.read_csv(r"results.csv")
df_result
```

```python
merge1=df_result.merge(driver,left_on='driverId',right_on='driverId')
merge1=merge1.drop(columns=['driverId'])
```

```python
merge2=merge1.merge(race,left_on='raceId',right_on='raceId')
merge2.columns
```

```python
merge3=merge2.drop(columns=['points','raceId','number','positionText','laps','time',
       'milliseconds','fastestLapTime','fastestLapSpeed','statusId','fastestLap','positionOrder'])
merge3
```

```python
df_con=pd.read_csv(r'constructors.csv')
con=df_con.iloc[:,0:2]
con
```

```python
merge4=merge3.merge(con,left_on='constructorId',right_on='constructorId')
result=merge4.drop(columns=['constructorId'])
result
```

```python
result.to_csv("clean_result.csv")
```
